```python
import pandas as pd
```

```python
Power = pd.read_csv("SolarHistory.csv", index_col="Date")
Power
```

```python
imp_Power = Power[["Yield(Wh)","Consumption(Wh)","Max. PV power(W)","Max. PV voltage(V)","Max. battery voltage(V)","Time in bulk(m)","Time in absorption(m)","Time in float(m)"]].copy()
```

```python
imp_Power
```

```python
imp_Power.dtypes
```

```python
imp_Power.index = pd.to_datetime(imp_Power.index)
imp_Power.index
```

```python
imp_Power[["Yield(Wh)"]].plot()
```

```python
imp_Power[["Max. PV power(W)"]].plot()
```

```python
imp_Power["target"] = imp_Power.shift(-1)["Yield(Wh)"]
```

```python
imp_Power
```

```python
imp_Power = imp_Power.iloc[:-1,:].copy()
imp_Power
```

```python
from sklearn.linear_model import Ridge

reg = Ridge(alpha=.1)
```

```python
predictors = ["Yield(Wh)", "Consumption(Wh)"]
```

```python
imp_Power = imp_Power.sort_index()
train = imp_Power.loc[:"2024-09-21"]
test = imp_Power.loc["2024-09-22":]
```

```python
reg.fit(train[predictors], train["target"])
```

```python
predictions = reg.predict(test[predictors])
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(test["target"], predictions)
```

```python
combined = pd.concat([test["target"], pd.Series(predictions, index = test.index)], axis =1)
combined.columns = ["actual", "predictions"]
```

```python
combined
```
